```python
import pandas as pd
```

### Reading data

```python
original_seed = pd.read_csv(
    "seed_matrices/1111708_original_seed.csv", sep="\t", header=None, names=["Protein", "GO ID", "Score"]
)
# convert scores to 1s
original_seed["Score"] = 1
original_seed.sort_values(by=["Protein"], inplace=True)
original_seed
```

```python
new_seed = pd.read_csv(
    "seed_matrices/83333_new_seed.tsv", sep="\t"
)
new_seed.drop(columns=["interpro"], inplace=True)
# Order columns by Protein column
new_seed.sort_values(by=["Protein"], inplace=True)
new_seed
```

### Intersecting new and old seeds

```python
# Intersecting new and old seeds only on GO ID column
go_id_old = set(original_seed["GO ID"].unique())
go_id_new = set(new_seed["GO ID"].unique())
common_go_ids = go_id_old.intersection(go_id_new)
print(f"Number of common GO IDs: {len(common_go_ids)}")
print(f"Number of GO IDs in original seed: {len(go_id_old)}")
print(f"Number of GO IDs in new seed: {len(go_id_new)}")
```
